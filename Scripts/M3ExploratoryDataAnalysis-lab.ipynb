Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
Q1 = df['ConvertedComp'].quantile(0.25)
Q3 = df['ConvertedComp'].quantile(0.75)

# Calculate the IQR by subtracting Q1 from Q3
IQR = Q3 - Q1
print(f"The first Quantile : {Q1}")
print(f"The third Quantile : {Q3}")
print(f"The Interquartile Range (IQR) for 'ConvertedComp' is: {IQR}")
```

Find out the upper and lower bounds.


```python
# your code goes here
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
(lower_bound, upper_bound)
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# Count outliers below the lower bound
outliers_below = df[df['ConvertedComp'] < lower_bound].shape[0]

# Count outliers above the upper bound
outliers_above = df[df['ConvertedComp'] > upper_bound].shape[0]

# Total number of outliers
total_outliers = outliers_below + outliers_above

total_outliers
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
df_no_outliers = df[(df['ConvertedComp'] >= lower_bound) & (df['ConvertedComp'] <= upper_bound)]

df_no_outliers
```
